```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movies_df = pd.read_csv("movies.csv")
```

```python
movies_df.drop('recommendations', axis=1, inplace=True)
```

```python
movies_df.shape
```

```python
movies_df.dropna(inplace=True)
```

```python
movies_df.shape
```

```python
movies_df.isna().sum()
```

```python
movies_df.duplicated()
```

```python
movies_df.duplicated().sum()
```

```python
movies_df.drop_duplicates(inplace=True)
```

```python
movies_df.dtypes
```

```python
movies_df['release_date'] = movies_df['release_date'].astype(object)
```

```python
all_genres = movies_df['genres'].dropna().str.split('-').sum()
```

```python
genres_df = pd.DataFrame(all_genres, columns=['genre'])
```

```python
genre_counts = genres_df['genre'].value_counts()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_counts.index, y=genre_counts.values, color="skyblue")
plt.title('Distribution of Movie Genres')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.show()
```

```python
movies_df['revenue'] = pd.to_numeric(movies_df['revenue'], errors='coerce')
```

```python
top_10_highest_grossing=movies_df.sort_values(by='revenue', ascending=False).head(10)
```

```python
print(top_10_highest_grossing[['title', 'revenue']])
```

```python
movies_df['budget'] = pd.to_numeric(movies_df['budget'], errors='coerce')
```

```python
correlation = movies_df['budget'].corr(movies_df['revenue'])
```

```python
print(f"Correlation between budget and revenue: {correlation}")
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=movies_df, x='budget', y='revenue', alpha=0.5)
plt.title('Relationship Between Movie Budget and Revenue')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.show()
```

```python
movies_df['release_date'] = pd.to_datetime(movies_df['release_date'], errors='coerce')
```

```python
movies_df['release_year'] = movies_df['release_date'].dt.year
```

```python
movies_per_year = movies_df['release_year'].value_counts().sort_index()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(movies_per_year.index, movies_per_year.values, marker='o', linestyle='-', color='b')
plt.title('Trend of Movie Production Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Movies Produced')
plt.grid(True)
plt.show()
```
